```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = np.loadtxt('datasets/apple_prices.csv')
```

```python
fig = plt.figure(figsize = (10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
agap = (max(a) - min(a))*0.1
ax1.plot(a,c = 'k')
ax1.set_xlim([0,len(a)])
ax1.set_ylim([min(a)-agap,max(a)+agap])
ax1.axis('off')
ax2.scatter(np.arange(len(a)),a,c = 'k',edgecolor = 'w',s = 30)
ax2.axis('off')
ax2.set_xlim([0,len(a)])
ax2.set_ylim([min(a)-agap,max(a)+agap])
```
